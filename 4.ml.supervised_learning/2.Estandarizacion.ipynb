# Estandarizacin 
La estandarizacin consiste en transformar la distribucin de los valores de los atributos para que estos posean una media y desviacin tpica determinadas. Esta transformacin se lleva a cabo para cada atributo de forma independiente.



A cada valor del atributo se le resta la media aritmtica de los valores originales del atributo y se divide por la desviacin tpica de estos. Al igual que la normalizacin, la operacin de resta es el centrado, que, en el caso de la estandarizacin, en lugar de centrarse en el mnimo (normalizacin), se centra en la media. 

El escalado en estandarizacin comprende tan solo la amplitud definida por la desviacin tpica. Por consiguiente, los valores estandarizados no estn acotados en un rango definido (como suceda con la normalizacin).  

Una vez un atributo est estandarizado, suele interesar desestandarizar valores a su distribucin original. 

Para calcular el valor `x` desestandarizado a partir del estadarizado, basta aplicar la operacin en sentido contrario.



```python
from sklearn import preprocessing
import numpy as np
```

En este programa se ha vuelto a hacer uso del paquete preprocessing de la librera scikit-learn. La estandarizacin se encuentra en la clase StandardScaler. 

```python
# Carga de datos.
X_train = np.array([[ 1., -1.,  2.],
                    [ 2.,  0.,  0.],
                    [ 0.,  1., -1.]])
```

```python
# Limpieza de datos: estandarizacin.
standardizer = preprocessing.StandardScaler() #Aplicamos el StandardScaler para poder estandarizar
X_train_std = standardizer.fit_transform(X_train)
```

Tal como se aprecia en el programa, tenemos la misma matriz de nmeros vista anteriormente y se ha creado un objeto llamado standardizer de la clase `StandardScaler`. Por ltimo, el mtodo `fit_transform()` nos crea el modelo de estandarizacin y lo aplica sobre la matriz de datos.

```python
# Salida de resultados.
print(X_train_std) #Se puede observar que los nmeros no caen entre 0 y 1 como en la normalizacin
```

```python
X_train_2 = np.array([[ 1., -1.,  2.],
                    [ 2.,  0.,  0.],
                    [ 0.,  1., -1.]])
```

```python
f = standardizer.fit(X_train_2)
X_train_std_2 = f.transform(X_train_2)
```

```python
print(X_train_std_2)
```

```python
# Comprobar si las matrices estandarizadas son iguales
print(np.array_equiv(X_train_std, X_train_std_2))
```
