# Seleccin de atributos (Feature Selection)

La seleccin de atributos es una subtarea dentro de la tarea de reduccin de dimensionalidad, la cual, a su vez, se incluye dentro de la `limpieza de datos`. 

Las tcnicas de seleccin de atributos pueden ser tanto `supervisadas` como `no supervisadas`, en funcin de si estas consideran o no la existencia de una clase en el conjunto de datos. 

Todas las tcnicas de seleccin de atributos necesitan de un algoritmo de evaluacin de atributos que les permita determinar la bondad de estos con el fin de seleccionar aquellos que puedan ser candidatos para ser eliminados. 

Existen numerosos algoritmos de evaluacin de atributos, dependiendo de cmo calculen la mtrica de bondad, pero existen dos categoras: algoritmos de evaluacin de atributos individuales (o univariantes) y algoritmos de evaluacin de conjuntos de atributos (o multivariantes). 

>Note: Ntese que la relevancia, o poder predictivo, que un atributo puede tener de cara a la prediccin de una clase puede depender de la existencia de otros atributos. Esto es, un atributo puede recibir una mtrica baja de bondad, por su dbil relacin matemtica con la variable clase y, sin embargo, ser de gran importancia cuando contribuye, junto a otros, a la prediccin de esta. Por esta razn, los algoritmos de evaluacin de conjuntos de atributos, aunque mucho ms costosos computacionalmente que los de atributos individuales, suelen seleccionar atributos que, en conjunto, son ms relevantes.

En el caso de la seleccin de atributos supervisada, el algoritmo de evaluacin puede estar definido solo para clasificacin o para regresin, o bien para ambos tipos de aprendizaje. Por ello, es importante tener en cuenta la aplicabilidad del algoritmo deseado al conjunto de datos y seleccionar uno que sea compatible.

Cuando se lleva a cabo una seleccin de atributos mediante un algoritmo de evaluacin de conjuntos de atributos, se necesita tambin disponer de un algoritmo de bsqueda, debido a que el espacio de posibles subconjuntos que hay que explorar es exponencial con respecto al nmero de atributos inicial del conjunto de datos.

## VarianceThreshold

El algoritmo `sklearn.feature_selection.VarianceThreshold` es una funcin de seleccin de caractersticas que elimina las caractersticas con una varianza muy baja. Es un algoritmo de seleccin de caractersticas no supervisado, lo que significa que no utiliza la variable objetivo para seleccionar las caractersticas.

El algoritmo funciona calculando la varianza de cada caracterstica en el conjunto de datos. La varianza es una medida de la dispersin de los datos alrededor de la media. Las caractersticas con una varianza muy baja tienen valores muy similares, por lo que no aportan mucha informacin sobre el conjunto de datos.

El algoritmo elimina las caractersticas con una varianza inferior a un cierto umbral. El umbral se puede establecer manualmente o se puede calcular automticamente utilizando un mtodo de validacin cruzada.

Por ejemplo, si tenemos un conjunto de datos con `100` muestras y `10` predictores, el algoritmo calcula la varianza de cada predictor. Si el umbral se establece en `0.1`, el algoritmo eliminar las caractersticas con una varianza inferior a `0.1`.

El algoritmo `sklearn.feature_selection.VarianceThreshold` es un algoritmo simple y eficiente que puede ser til para eliminar caractersticas irrelevantes de un conjunto de datos. Sin embargo, es importante tener en cuenta que este algoritmo no tiene en cuenta la relacin entre las caractersticas y la variable objetivo. Por lo tanto, es posible que elimine caractersticas que son relevantes para el objetivo.

Aqu hay un resumen de las ventajas y desventajas del algoritmo `sklearn.feature_selection.VarianceThreshold`:

Ventajas:

- Es simple y eficiente.
- No requiere la variable objetivo.

Desventajas:

- Puede eliminar caractersticas relevantes para el objetivo.
- No tiene en cuenta la interaccin entre las caractersticas.
- En general, el algoritmo sklearn.feature_selection.VarianceThreshold es una buena opcin - para eliminar caractersticas irrelevantes de un conjunto de datos. 

## Ejemplo

En el Programa se construye una tabla de datos con tres atributos, de los cuales el primero es prcticamente constante (salvo un valor a 1, el resto 5 de 6 veces vale 0). La clase que implementa este tipo de seleccin en scikit-learn es `VarianceThreshold`. El umbral utilizado en este ejemplo es 0,8, el cual deja pasar un atributo si vara al menos un 20 % a travs de las instancias. Tras la aplicacin de la seleccin, el conjunto de datos resultante solo contiene los dos ltimos atributos 

```python
import numpy as np
from sklearn.feature_selection import VarianceThreshold
```

```python
# Carga de datos.
X = [[0, 0, 1], [0, 1, 0], [1, 0, 0],
     [0, 1, 1], [0, 1, 0], [0, 1, 1]]


print(np.array(X))
```

```python
# Limpieza de datos: seleccin de atributos.

# fijamos un umbral (th) para decidir qu porcentaje mnimo deben variar los valores 
# de una columna para considerarse una columna relevante a la hora de 
# mantenerla en nuestro estudio
th = 0.2
sel = VarianceThreshold(th)
print(sel.fit_transform(X))
```
